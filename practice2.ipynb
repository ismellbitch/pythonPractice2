```python

class Task:
    def __init__(self, hours, minutes, seconds):
        self.hours = hours
        self.minutes = minutes
        self.seconds = seconds
    def count_seconds(self):
        all=0
        print("Введите часы: ", end=" ")
        h = int(input())
        all+=h*3600
        print("Введите минуты: ", end=" ")
        m = int(input())
        all+=m*60
        print("Введите секунды: ", end=" ")
        s = int(input())
        all+=s
        return all
    def add_seconds(self):
        if self.seconds < 55:
            self.seconds+=5
        else:
            self.seconds += 5 - 60
            if self.minutes < 59:
                self.minutes+=1
            else:
                self.minutes = 0
                self.hours += 1


class Train(Task):
    def __init__(self, hours, minutes, seconds, train_number, direction, departure_hours, departure_minutes, departure_seconds):
        super().__init__(hours, minutes, seconds)
        self.train_number = train_number
        self.direction = direction
        self.departure_hours = departure_hours
        self.departure_minutes = departure_minutes
        self.departure_seconds = departure_seconds
    def time_before_departure(self):
        if self.departure_hours <= self.hours and self.departure_minutes <= self.minutes and self.departure_seconds <= self.seconds:
            return print("Поезд уже отправился")
        else:
            hours_last = self.departure_hours - self.hours
            minutes_last = self.departure_minutes - self.minutes
            seconds_last = self.departure_seconds - self.seconds
            print("Поезд отправляется через ", hours_last," часов, ", minutes_last, " минут и ", seconds_last, " секунд")


class Cable:
    def __init__(self, type, cores_number, diameter):
        self.type = type
        self.cores_number = cores_number
        self.diameter = diameter
    def getQ(self):
        q = self.diameter / self.cores_number
        return q

class CableExt(Cable):
    def __init__(self, type, cores_number, diameter, p):
        super().__init__(type, cores_number, diameter)
        self.p = True
    def getQ(self):
        q = self.diameter / self.p
        return q
    def getQP(self):
        if self.p == True:
            self.qp = 2 * self.getQ()
            return self.qp
        else:
            self.qp = 0.7 * self.getQ()
            return self.qp

print("Задание 1")

obj1 = Task(5, 59, 59)
print("до:")
print(obj1.hours)
print(obj1.minutes)
print(obj1.seconds)
print("после:")
obj1.add_seconds()
print(obj1.hours)
print(obj1.minutes)
print(obj1.seconds)

print("Задание 2")

obj2 = Train(5, 59, 59, 1, "left", 5, 59, 59)
obj2.time_before_departure()

print("Задание 3")

cab = CableExt(1, 12, 5, True)
print(cab.getQP())
```
